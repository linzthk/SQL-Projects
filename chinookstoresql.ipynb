Which genre sells the most in the United States? Rock Genre with 561 tracks sold. Solved by joining customer, invoice line and invoice. 

```python
%%sql

WITH usa_tracks_sold AS
   (
    SELECT * FROM invoice_line xx
    INNER JOIN invoice yy on xx.invoice_id = yy.invoice_id
    INNER JOIN customer zz on yy.customer_id = zz.customer_id
    WHERE zz.country = "USA"
   )

SELECT
    x.name genre,
    count(uts.invoice_line_id)tracks_sold ,
    cast(count(uts.invoice_line_id) AS FLOAT) / (
        SELECT COUNT(*) from usa_tracks_sold
    ) percentage_sold
FROM usa_tracks_sold uts
INNER JOIN track t on t.track_id = uts.track_id
INNER JOIN genre x on x.genre_id = t.genre_id
GROUP BY 1
ORDER BY 2 DESC
LIMIT 10;
```

Queried the database to see which employees are generating the most sales.

```python
%%sql
SELECT
    employee.first_name || ' ' || employee.last_name AS sales_representative,
    employee.hire_date AS hire_date,
    employee.city AS work_location,
    ROUND(SUM(i.total), 2) AS total_sales
  FROM invoice i
 INNER JOIN customer c ON c.customer_id = i.customer_id
 INNER JOIN employee ON employee.employee_id = c.support_rep_id
 GROUP BY sales_representative
 ORDER BY total_sales DESC;
```
